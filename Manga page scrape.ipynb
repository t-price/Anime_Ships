# Manga data

### Importing the needed libraries

```python
from bs4 import BeautifulSoup
import time
import urllib.request
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_My_Hero_Academia_chapters'

with urllib.request.urlopen(url) as response:
    html = response.read()
    html = html.decode('utf-8')
    
with open ('manga.html', 'w') as new_file:
    new_file.write(html)
```

```python
manga_soup = BeautifulSoup(html, 'html.parser')
manga_soup.prettify
```

```python
tables = manga_soup.find('table')
```

```python
table = pd.read_html(str(tables))[0]
table.head(20)


#table.to_csv(r'C:\Users\tessa\Anime_ships\mangas_notclean.csv',index=False, header= True)
#table
```

```python
mangas = table.iloc[::5]
mangas
```

```python

original_release = mangas["Original release date"].to_string()
#print(type(original_release))
#original_release
original_split = mangas["Original release date"].str.split(pat='[',expand=True)
original = original_split[0].to_frame()
original.columns = ["Original_Release_Date"]

original

mangas_df = mangas.join(original)
```

```python

english_release = mangas["English release date"].to_string()
#print(type(english_release))
#english_release
english_split = mangas["English release date"].str.split(pat='[',expand=True)
english = english_split[0].to_frame()
english.columns = ["English_Release_Date"]

english

mangas_dates = mangas_df.join(english)

mangas_dates
```

```python
import datetime 

mangas_dates["Original_Release_Date"] = pd.to_datetime(mangas_dates["Original_Release_Date"])
mangas_dates

mangas_dates["English_Release_Date"] = pd.to_datetime(mangas_dates["English_Release_Date"], errors='coerce')
mangas_dates
```
