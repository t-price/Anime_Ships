# Anime dates clean up

### Importing the needed libraries

```python
from bs4 import BeautifulSoup
import time
import urllib.request
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_My_Hero_Academia_episodes'

with urllib.request.urlopen(url) as response:
    html = response.read()
    html = html.decode('utf-8')
    
with open ('anime.html', 'w') as new_file:
    new_file.write(html)
```

```python
anime_soup = BeautifulSoup(html, 'html.parser')
anime_soup.prettify
```

```python
tables = anime_soup.find_all('table')
```

```python
table_df = []
for x in range(len(tables)):
    table_df.append(pd.read_html(str(tables))[x])
```
