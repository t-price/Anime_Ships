```python
import pandas as pd

table = pd.read_csv(r'C:\Users\tessa\Anime_ships\mangas_notclean.csv')
table.head(20)
```

```python
mangas = table.iloc[::5]
mangas
```

```python

original_release = mangas["Original release date"].to_string()
#print(type(original_release))
#original_release
original_split = mangas["Original release date"].str.split(pat='[',expand=True)
original = original_split[0].to_frame()
original.columns = ["Original_Release_Date"]

original

mangas_df = mangas.join(original)
mangas_df
```

```python

english_release = mangas["English release date"].to_string()
#print(type(english_release))
#english_release
english_split = mangas["English release date"].str.split(pat='[',expand=True)
english = english_split[0].to_frame()
english.columns = ["English_Release_Date"]

english

mangas_dates = mangas_df.join(english)

mangas_dates
```

```python
import datetime 

mangas_dates["Original_Release_Date"] = pd.to_datetime(mangas_dates["Original_Release_Date"])
mangas_dates

mangas_dates["English_Release_Date"] = pd.to_datetime(mangas_dates["English_Release_Date"], errors='coerce')
mangas_dates
```

```python
mangas_dates.to_csv(r'C:\Users\tessa\Anime_ships\mangas_clean.csv',index=False, header= True)
print(mangas_dates)
```
